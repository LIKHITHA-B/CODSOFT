```python
import pandas as pd
import numpy as np
```

```python
datasix1 = pd.read_csv('/content/GP1_0.6_marker.csv')
dataseven2 = pd.read_csv('/content/GP1_0.7_marker.csv')
datasix1.head()
```

```python
data1 = pd.DataFrame({'secim':np.zeros(12000)})
data2 = pd.DataFrame({'secim':np.ones(12000)})

data2 = pd.concat([dataseven2,data2],axis=1)
data1 = pd.concat([datasix1,data1],axis=1)

data = data1.append(data2,ignore_index=True)
data.head()
```

```python
x = data.drop(["secim"],axis=1)
y = data.secim.values
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.svm import SVC

# Assuming x_train, y_train, x_test, y_test are already defined

# Create and train the SVM model
svm = SVC(random_state=1)
svm.fit(x_train, y_train)

# Make predictions on the test set
y_pred = svm.predict(x_test)

# Calculate accuracy in percentage
accuracy = accuracy_score(y_test, y_pred) * 100
print("Accuracy of SVM: {:.2f}%".format(accuracy))

# Create a confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Visualize the confusion matrix using seaborn heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False,
            xticklabels=["Class 0", "Class 1"], yticklabels=["Class 0", "Class 1"])
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import make_classification

# Assuming x_train, y_train are already defined
# Generating a simple dataset for illustration purposes
# Replace the following two lines with your actual data
x_train, y_train = make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0, random_state=1)

# Create and train the SVM model
svm = SVC(kernel='linear', random_state=1)
svm.fit(x_train, y_train)

# Plot the data points
plt.scatter(x_train[:, 0], x_train[:, 1], c=y_train, cmap=plt.cm.Paired, edgecolors='k', marker='o', s=70, label="Data")

# Plot the decision boundary lines
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# Create grid to evaluate model
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 50), np.linspace(ylim[0], ylim[1], 50))
Z = svm.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot decision boundary and margins
ax.contour(xx, yy, Z, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'], linewidths=[2, 2, 2])

# Highlight support vectors
ax.scatter(svm.support_vectors_[:, 0], svm.support_vectors_[:, 1], s=100, facecolors='none', edgecolors='k', marker='s', label='Support Vectors')

plt.title("SVM Decision Boundary")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```
